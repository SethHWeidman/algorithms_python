```python
with open("clustering1.txt", "rb") as f:
    cluster_list = [x.decode('utf-8').strip('\n') for x in f.readlines()[1:]]
```

```python
cluster_list_clean = [list(map(int, x.split())) for x in cluster_list]
```

```python
len(cluster_list_clean)
```

```python
from collections import defaultdict

big_graph = defaultdict(dict)
for edge in cluster_list_clean:
    big_graph[edge[0]][edge[1]] = edge[2]
    big_graph[edge[1]][edge[0]] = edge[2]
```

`edges`: a list where first element is a tuple, second element is the distance; sort by second element:

```python
edges = []
for edge in cluster_list_clean:
    edges.append(((edge[0], edge[1]), edge[2]))
```

```python
edges.sort(key=lambda x: x[1])  
```

```python
edges
```

```python
leaders = defaultdict(int)
for edge in cluster_list_clean:
    leaders[edge[0]] = edge[0]
```

```python
leaders
```

Strategy:

1. Find minimum edge (takes O(m)) that is not connected
2. Using a union find data structure, merge clusters and update leaders
3. Keep track of total number of clusters
