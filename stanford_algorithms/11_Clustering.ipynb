Strategy:

1. Find minimum edge (takes O(m)) that is not connected
2. Using a union find data structure, merge clusters and update leaders
3. Keep track of total number of clusters

```python
edges = []
for edge in cluster_list_clean:
    edges.append(((edge[0], edge[1]), edge[2]))
    
node_leaders = defaultdict(int)
for edge in cluster_list_clean:
    node_leaders[edge[0]] = edge[0]

leader_followers = defaultdict(list)
for edge in cluster_list_clean:
    leader_followers[edge[0]] = [edge[0]]
```

```python
def find_min_edge(edges):
    edges.sort(key=lambda x: x[1]) 
    min_edge = edges.pop(0)
    return min_edge
```

```python
def num_clusters(leaders):
    return len(list(set(leaders.values())))
```

```python
def find_cluster_size(node):
    return len(leader_followers[node_leaders[node]])
```

```python
while num_clusters(node_leaders) >= 4:
    print(num_clusters(node_leaders))
    min_edge = find_min_edge(edges)
    node1, node2 = min_edge[0][0], min_edge[0][1]
#     import pdb; pdb.set_trace()
    if find_cluster_size(node1) >= find_cluster_size(node2):
        leader_followers[node1] = leader_followers[node1] + leader_followers[node2]
        for node in leader_followers[node2]:
            node_leaders[node] = node_leaders[node1]
        leader_followers[node2] = []
    else:
        leader_followers[node2] = leader_followers[node1] + leader_followers[node2]
        for node in leader_followers[node1]:
            node_leaders[node] = node_leaders[node2]
        leader_followers[node1] = []
        
        
```

```python
NUMLIST_FILENAME = "clustering1.txt" # result: 106
# NUMLIST_FILENAME = "data/tests/msk-clustering-1.1.txt" # result: 2
# NUMLIST_FILENAME = "data/tests/msk-clustering-1.2.txt" # result: 5

inFile = open(NUMLIST_FILENAME, 'r')

graph = []
merged_nodes = {} 
# merged_nodes stores nodes as keys, cluster number as values
clusters_merged = 0
num_clusters = 0
k = 4

for f in inFile:
    if(num_clusters == 0):
        num_clusters = int(f.strip())
    else:
        node1, node2, edge_cost = map(int, f.split())
        graph.append([node1, node2, edge_cost])

# sorting graph by increasing order of edge cost
graph = sorted(graph, key=lambda x: x[2])
```

```python
index = 0

while num_clusters > k:
    merged = False

    # graph has already been sorted: shortest nodes at the front 
    # merge of single node to a clusters
    if graph[index][0] not in merged_nodes:
        # if the node1 is not already in merged_nodes...
        if graph[index][1] not in merged_nodes:
            # ...and neither is node2, create a new cluster, starting with cluster 1
            clusters_merged += 1
            cluster = clusters_merged
        else:
            # ...if node2 is already in merged_nodes, then assign "cluster" to be the cluster of node2 
            cluster = merged_nodes[graph[index][1]]
        # Assign node1 to the appropriate cluster
        merged_nodes[graph[index][0]] = cluster
        # If node1 was not in merged_nodes, but it now is, set merged=True
        merged = True

    # This block ensures node2 also gets assigned to the correct cluster
    # If node2 is not in merged_nodes...
    # merge of single node to a clusters
    if graph[index][1] not in merged_nodes:
        # And neither is node1...
        if graph[index][0] not in merged_nodes:
            # And we have not merged yet, which means:
            # ?
            if not merged:
                clusters_merged += 1
            cluster = clusters_merged
        else:
            # This line, and the cluster assignment line above it, ensure that we assign node2 to the right 
            # cluster, either:
            # a new cluster
            # The cluster of node1
            cluster = merged_nodes[graph[index][0]]
        # Do the assignment
        merged_nodes[graph[index][1]] = cluster 
        merged = True
    
    nodes_in_merged_nodes = graph[index][0] in merged_nodes and graph[index][1] in merged_nodes

    # inner merge of clusters with multiple nodes
    # separate merge_cluster function
    if nodes_in_merged_nodes and merged_nodes[graph[index][0]] != merged_nodes[graph[index][1]]:
        min_cluster = min(merged_nodes[graph[index][0]], merged_nodes[graph[index][1]])
        max_cluster = max(merged_nodes[graph[index][0]], merged_nodes[graph[index][1]])
        for key in merged_nodes:
            if merged_nodes[key] == max_cluster:
                merged_nodes[key] = min_cluster
                merged = True

    # removing the edge
    del graph[index]

    # If there has been a merge:
    # Clean up the graph
    # Decrement "num_clusters"
    if merged:
        idx = 0
        while len(graph) > idx:
            nodes_in_merged_nodes = graph[idx][0] in merged_nodes and graph[idx][1] in merged_nodes
            if nodes_in_merged_nodes and merged_nodes[graph[idx][0]] == merged_nodes[graph[idx][1]]:
                # removing inner cluster edges
                del graph[idx]
            else:
                idx += 1
        num_clusters -= 1

# result is the distance of the smallest edge after removing all redundant inner cluster edges and merging
# all clusters until reaching k number of clusters
print('result: ' + str(graph[0][2]))
```
